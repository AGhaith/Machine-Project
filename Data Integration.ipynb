```python
import pandas as pd
import glob

all_files = glob.glob("./Sources Before Integration/*.csv")

df = pd.concat((pd.read_csv(f) for f in all_files), ignore_index=True)
df.to_csv("combined_data.csv", index=False)
print("Total records:", len(df))
print("Columns:", df.columns.tolist())
```

```python
# Drop columns missing more than 50% of their values
old = df.columns.size
df = df.dropna(axis=1, thresh=len(df) * 0.5)
new = df.columns.size
print("dropped", (old - new), " missing values by dropping columns with >50% missing data")

# Fill the rest with column means (numeric only)
df = df.fillna(df.mean(numeric_only=True))

# Check remaining missing values
print("Remaining missing values:", df.isna().sum().sum())
```

```python
# For any remaining missing values:
df = df.fillna(method='ffill').fillna(method='bfill')

# Confirm
print("Remaining missing values:", df.isna().sum().sum())
```
